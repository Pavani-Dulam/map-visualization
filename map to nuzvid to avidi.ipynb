```python
import folium
```

```python
way1=[[16.7850,80.8488],[16.6356,80.9716],[16.7625,81.8423],[16.6727,81.9030]]
way2=[[16.7850,80.8488],[17.0005, 81.8040],[16.6727,81.9030]]
```

```python
from branca.element import Figure
fig=Figure(width=550,height=350)
```

```python
f1=folium.FeatureGroup("transportation way1")
f2=folium.FeatureGroup("transportation way2")
map=folium.Map(location=[16.7850,80.8488],title='cartodbpositron')
line_1=folium.vector_layers.PolyLine(way1,popup='<b>Path of Vehicle_1</b>',tooltip='Vehicle_1',color='green',weight=10).add_to(f1)
line_2=folium.vector_layers.PolyLine(way2,popup='<b>Path of Vehicle_1</b>',tooltip='Vehicle_1',color='red',weight=10).add_to(f2)
f1.add_to(map)
f2.add_to(map)
folium.Marker(location=[16.7850,80.8488],popup='starting location',tooltip='Nuzvid',icon=folium.Icon(color='purple',prefix='fa',icon='bus')).add_to(map)
folium.Marker(location=[16.6356,80.9716],popup='2nd location',tooltip='Hanuman Junction',icon=folium.Icon(color='red',prefix='fa',icon='bell')).add_to(map)
folium.Marker(location=[16.7625,81.8423],popup='3rd location',tooltip='Ravulapalem',icon=folium.Icon(color='blue',prefix='fa',icon='stethoscope')).add_to(map)
folium.Marker(location=[16.6727,81.9030],popup='4th location',tooltip='Avidi',icon=folium.Icon(color='green',prefix='fa',icon='tree')).add_to(map)

#way2 markers
folium.Marker(location=[17.0005,81.8040],popup='2nd location',tooltip='rajuhmundry',icon=folium.Icon(color='orange',prefix='fa',icon='train')).add_to(map)
folium.LayerControl().add_to(map)
map
```
